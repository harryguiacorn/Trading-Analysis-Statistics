```python
import yfinance as yf
import ta
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download historical data for desired ticker symbol
data = yf.download('AAPL', period="max")

# Calculate RSI
data['RSI'] = ta.momentum.rsi(data['Close'])

# Save OHLCV and RSI data to a CSV file
data.to_csv('AAPL_data_with_RSI.csv')

# Create a bell curve chart of the RSI data
plt.figure(figsize=(10,6))
sns.histplot(data['RSI'].dropna(), bins=100, color='blue', kde=True)
# sns.kdeplot(data['RSI'].dropna(), color='red')
plt.title('Distribution of RSI values')
plt.show()
```
